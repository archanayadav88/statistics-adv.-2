**Question 1: What is a random variable in probability theory?**

**ANS:**A random variable is a key concept in statistics that connects theoretical probability with real-world data. It is a function that assigns a real number to each outcome in the sample space of a random experiment.

#Random Variable is expressed as:-
            X: S →R
            
where:              
-X is a Random Variable (It is usually denoted using a capital letter)
-S is Sample Space
-R is a Set of Real Numbers  

**For example:**

When you roll a die, the outcome is one of the six faces. A random variable can assign a number (like 1 to 6) to each of these outcomes, allowing us to analyze the results using statistical methods.

**Question 4: What is a binomial distribution, and how is it used in probability**?

**ANS:**A Binomial Distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes — success or failure.

**Conditions for a Binomial Distribution:**

-A random variable X follows a binomial distribution if:

1)There are a fixed number of trials (n).

2)Each trial has only two outcomes — success or failure.

3)The probability of success (p) is the same for each trial.

4)All trials are independent of each other.



**Question 5: What is the standard normal distribution, and why is it important?**

**ANS:**Standard normal distribution, also known as the z-distribution, is a special type of normal distribution. In this distribution, the mean (average) is 0 and the standard deviation (a measure of spread) is 1. This creates a bell-shaped curve that is symmetrical around the mean.

The standard normal distribution is a special type of normal distribution that has:

Mean (μ) = 0

Standard deviation (σ) = 1

It is often represented by the letter Z, so it is also called the Z-distribution.

#Importance of the Standard Normal Distribution:

1)Basis for Z-scores:

-Any normal distribution can be converted into a standard normal distribution using a Z-score:
                  Z=X-μ​/σ


This allows comparison between different datasets.

2)Simplifies Probability Calculations:

-Standard normal tables (Z-tables) provide probabilities for Z-values, which help find areas under the curve (probabilities).

3)Widely Used in Statistics:

-It's essential for hypothesis testing, confidence intervals, and control charts.

4)Helps in Data Standardization:

-Converting data to Z-scores helps in identifying outliers and comparing scores from different distributions.

5)Foundation for Many Models:

-Many statistical and machine learning algorithms assume or are based on normal or standard normal distribution behavior.

Example:

If test scores have a mean of 70 and a standard deviation of 10, a score of 85 can be converted to a Z-score:

Z=85-70/10 = 1.5

This means the student scored "1.5" standard deviations above the mean, and we can use the Z-table to find their percentile.

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

**ANS:**The Central Limit Theorem (CLT) states that when you take a large number of random samples from any population (with a finite mean and variance), the sampling distribution of the sample mean will approximate a normal distribution, regardless of the original population's shape.

#Why the CLT Is Critical in Statistics:

1)Foundation for Inferential Statistics:

-CLT allows statisticians to make inferences about population parameters (like mean or proportion) using sample data.

2)Enables Use of the Normal Distribution:

-Even if the population is not normal, the CLT justifies using Z-tests, t-tests, and confidence intervals based on the normal curve.

3)Simplifies Analysis:

-It makes complex, non-normal data manageable and predictable through the properties of the normal distribution.

4)Supports Probability Estimation:

-Helps in estimating probabilities of sample means, which is essential in hypothesis testing and quality control.

5)Used in Real-world Applications:

-Fields like finance, manufacturing, and data science rely on CLT for risk analysis, forecasting, and process optimization.

**Example:**

Suppose you're measuring the average height of 1,000 people.
Even if the population of heights is slightly skewed, the distribution of the sample means (for samples of, say, 50 people) will look approximately normal — allowing you to apply normal distribution methods for analysis.


**Question 7: What is the significance of confidence intervals in statistical analysis?**

**ANS:**
Significance of Confidence Intervals in Statistical Analysis:-
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence.

It is expressed as:

Confidence Interval=Sample Estimate ± Margin of Error

For example, a 95% confidence interval for a mean might look like:

                      (65.2,74.8)

This means we are 95% confident that the true population mean lies between 65.2 and 74.8.

#Significance

1)Provides a Range of Plausible Values:

-Instead of giving a single point estimate (like a mean of 70), a CI gives a range, showing the uncertainty around the estimate.

2)Reflects Reliability of Estimates:

-A narrow interval indicates a more precise estimate, while a wide interval suggests more variability or less certainty in the data.

3)Supports Decision Making:

-Confidence intervals help researchers and policymakers assess risk and reliability when making conclusions based on sample data.

4)Used in Hypothesis Testing:

-If a hypothesized value (like a population mean) falls outside the CI, it suggests that the hypothesis may be rejected.

5)Helps Compare Groups:

-Overlapping or non-overlapping confidence intervals can indicate whether differences between groups are statistically significant.

6)Communicates Statistical Uncertainty:

-CIs provide a clear way to express how confident we are about the sample's ability to represent the population.

**Example:**

If a survey finds that the average satisfaction score is 8.2 with a 95% CI of (7.8, 8.6), it means:

We are 95% confident that the true average satisfaction in the population lies between 7.8 and 8.6.

The result is statistically reliable within this range.



**Question 8: What is the concept of expected value in a probability distribution?**

**ANS:**
#Concept of Expected Value in a Probability Distribution

The expected value (EV) of a random variable is the long-run average or mean value it would take after many repetitions of a random experiment.
It represents the theoretical average outcome based on the probabilities of all possible values.

Example 1 :-(Discrete Case):

A fair die is rolled once.

Possible outcomes: 1, 2, 3, 4, 5, 6

Each has probability

            E(X)=(1+2+3+4+5+6)* 1/6 = 3.5

So, the expected value of rolling a fair die is 3.5 — the average outcome over many rolls.

Example 2:- (Weighted Case):

Suppose you play a game where:

Win ₹100 with probability 0.3

Lose ₹50 with probability 0.7

E(X)=(100)(0.3)+(-50)(0.7)=30-35=-5

This means, on average, you lose ₹5 per game in the long run.

#Significance of Expected Value:

1)Represents the Mean of a Probability Distribution:
-It gives a single summary measure of the central tendency.

2)Used in Decision-Making:
-Businesses, investors, and insurers use EV to evaluate risks and potential returns.

3)Foundation for Many Statistical Measures:
-Variance, standard deviation, and other metrics are derived using expected value.

4)Helps Compare Different Scenarios:
-Useful for choosing the most favorable or least risky option among several probabilistic outcomes.

**Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. **

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.

● Write the Python code to compute the mean sales and its confidence interval.

**ANS:**
The Central Limit Theorem (CLT) states that if we take repeated random samples from a population, the distribution of the sample means will be approximately normal, even if the population itself is not normal—as long as the sample size is large enough.

**In this case:**

We have sample sales data for 20 days (daily_sales).

We can use the sample mean and sample standard deviation to estimate the population mean.

The 95% confidence interval (CI) provides a range where the true average daily sales likely falls.

**Interpretation:**

-We are 95% confident that the true average daily sales for the company lie between ₹241.81 and ₹257.69.
This estimate is based on the Central Limit Theorem, which allows us to use sample data to infer the population mean.

```python
import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]


mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)


standard_error = std_sales / np.sqrt(n)


confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * standard_error
confidence_interval = (mean_sales - margin_of_error, mean_sales + margin_of_error)


print("Sample Mean Sales:", round(mean_sales, 2))
print("Sample Standard Deviation:", round(std_sales, 2))
print("95% Confidence Interval for Mean Sales:", confidence_interval)

```
